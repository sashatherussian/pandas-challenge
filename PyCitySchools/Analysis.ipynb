```python
import pandas as pd
import numpy as np
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete
```

```python
total_schools=school_data_complete['school_name'].nunique()
total_schools
```

```python
num_students = school_data_complete['Student ID'].nunique()
num_students
```

```python
total_budget=school_data['budget'].sum()
total_budget
```

```python
avg_math_score=school_data_complete['math_score'].mean()
avg_math_score
```

```python
avg_reading_score=school_data_complete['reading_score'].mean()
avg_reading_score
```

```python
percent_passing_math=len(school_data_complete[school_data_complete['math_score']>=70])/num_students
percent_passing_math
```

```python
percent_passing_reading=len(school_data_complete[school_data_complete['reading_score']>=70])/num_students
percent_passing_reading
```

```python
overall_passing=len(school_data_complete[(school_data_complete['math_score']>=70)&(school_data_complete['reading_score']>=70)])/num_students
overall_passing
```

```python
District_Sum=pd.DataFrame({'Total Schools':[total_schools],'Total Students':[num_students],
                           'Total Budget':[total_budget],'Average Math Score':[avg_math_score],
                           'Average Reading Score':[avg_reading_score],'% Passing Math':[percent_passing_math],
                           '% Passing Reading':[percent_passing_reading],'% Overall Passing':[overall_passing]})
District_Sum
```

```python
school_data['per student budget']=school_data['budget']/school_data['size']
school_data
```

```python
averages=school_data_complete.groupby('school_name')[['math_score','reading_score']].mean()
averages
```

```python
averages=averages.reset_index()
averages
```

```python
school_data=school_data.reset_index()
school_data
```

```python
school_data=pd.merge(school_data, averages)
```

```python
school_data_complete['math_passing']=np.where(school_data_complete['math_score']>=70,1,0)
school_data_complete['reading_passing']=np.where(school_data_complete['reading_score']>=70,1,0)
school_data_complete
```

```python
math_passing_num=school_data_complete.groupby('school_name',as_index=False)['math_passing'].sum()
math_passing_num
```

```python
school_data=pd.merge(school_data,math_passing_num)
school_data['percent passing math']=school_data['math_passing']/school_data['size']
school_data=school_data.drop('math_passing', axis=1)
school_data
```

```python
reading_passing_num=school_data_complete.groupby('school_name',as_index=False)['reading_passing'].sum()
reading_passing_num
```

```python
school_data=pd.merge(school_data,reading_passing_num)
school_data['percent passing reading']=school_data['reading_passing']/school_data['size']
school_data=school_data.drop('reading_passing', axis=1)
school_data
```

```python
school_data=school_data.drop('index',axis=1)
school_data=school_data.rename(columns={'school_name':'School Name'})
```

```python
school_data=school_data.rename(columns={'type':'School Type','size':'Total Students','budget':'Total School Budget',
                                        'per student budget':'Per Student Budget',
                                        'percent passing math':'% Passing Math',
                                        'percent passing reading':'% Passing Reading',
                                       'math_score':'Average Math Score','reading_score':'Average Reading Score'})
school_data
```

```python
school_data_complete['overall_passing']=np.where((school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70),1,0)
school_data_complete
```

```python
overall_passing_num=school_data_complete.groupby('school_name',as_index=False)['overall_passing'].sum()
overall_passing_num
```

```python
overall_passing_num=overall_passing_num.rename(columns={'school_name':'School Name'})

```

```python
school_data=pd.merge(school_data,overall_passing_num)
school_data['% Overall Passing']=school_data['overall_passing']/school_data['Total Students']
school_data=school_data.drop('overall_passing', axis=1)
school_data
```

```python
top_performing=school_data.sort_values(by='% Overall Passing', ascending=False)
top_performing.head(5)
```

```python
bottom_performing=school_data.sort_values(by='% Overall Passing')
bottom_performing.head(5)
```
